```python
x=int(input("Jumlah Pinjaman = Rp."))
tn=int(input("Lama Pinjaman = "))
print("=" * 30)
dp=x*(0.25)
bg=(x-dp)*((0.09)/12)
ang_pok=(x-dp)/(tn*12)
sa=x-dp
print("Harga Mobil      = Rp,",x)
print("Uang Muka        = Rp.", dp)
print("Flatpon Pinjaman = Rp.", sa )
print("=" * 30)
print("Angsuran Pokok /Bln = Rp.", ang_pok)
print("=" * 30)
print("Bunga /Bln = Rp.", bg)
print("=" * 30)
print("Angsuran /Bln = Rp.", ang_pok+bg)
print ("No : Angsuran Pokok  : Bunga     : Angsuran / Bln")
for i in range (1,((tn*12)+1)):
    print("="* 65)
    print(i, " : ", ang_pok,"      + ",bg," =", ang_pok+bg)
```

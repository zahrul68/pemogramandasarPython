```python
class Karyawan:
    def __init__(self, name, jabatan):
        self.gapok = 0
        self.gator = 0
        self.pajak = 0
        self.potongan = 0
        self.gaber = 0
        self.tunjangan = 0
        self.name = name
        self.jabatan = jabatan

class MasaKerjaGol(Karyawan):
    def __init__(self, name, jabatan, mkg, gol, ruang):
        super().__init__(name, jabatan)
        self.mkg = [0, 10, 20, 32]
        self.gol = ["I, II, III, IV"]
        self.ruang = ["a, b, c, d, e"]

class Gaji(MasaKerjaGol):
    def __init__(self, name, jabatan, mkg, gol, ruang):
        super().__init__(name, jabatan, mkg, gol, ruang)
        self.name = name
        self.jabatan = jabatan
        self.mkg = mkg
        self.gol = gol
        self.ruang = ruang

        if(jabatan == "Ahli Utama"):
            self.tunjangan += 1500000
        elif(jabatan == "Ahli Madya"):
            self.tunjangan += 1260000
        elif(jabatan == "Ahli Muda"):
            self.tunjangan += 960000
        elif(jabatan == "Ahli Pertama"):
            self.tunjangan += 540000

        if(mkg == 0):
            if(gol == "III"):
                if(ruang == "a"):
                    self.gapok += 2579400
                    self.gator += int(self.gapok) + int(self.tunjangan)
                    self.pajak += int(self.gapok * 0.05)
                    self.potongan += int(self.gapok * 0.1)
        elif(mkg == 10):
            if(gol == "III"):
                if(ruang == "a"):
                    self.gapok += 3012000
                    self.gator += int(self.gapok) + int(self.tunjangan)
                    self.pajak += int(self.gapok * 0.05)
                    self.potongan += int(self.gapok * 0.1)
        elif(mkg == 20):
            if(gol == "III"):
                if(ruang == "a"):
                    self.gapok += 3517200
                    self.gator += int(self.gapok) + int(self.tunjangan)
                    self.pajak += int(self.gapok * 0.05)
                    self.potongan += int(self.gapok * 0.1)
        elif(mkg == 32):
            if(gol == "III"):
                if(ruang == "a"):
                    self.gapok += 4235400
                    self.gator += int(self.gapok) + int(self.tunjangan)
                    self.pajak += int(self.gapok * 0.05)
                    self.potongan += int(self.gapok * 0.1)

    def HitungGaji(self):
        self.gaber += int(self.gator) - int(self.pajak) - int(self.potongan)

    def show_gaji(self):
        print("\nNama: {} \nJabatan: {}\nGolongan: {}/{} \nMasa Kerja Gol: {} Tahun \nGaji Pokok: Rp.{}\nTunjangan: Rp.{} \nPajak: Rp.{} \nPotongan: Rp.{} \nGaji Kotor: Rp.{} \nGaji Bersih: Rp.{}".format(
              self.name,
              self.jabatan,
              self.gol,
              self.ruang,
              self.mkg,
              self.gapok,
              self.tunjangan,
              self.pajak,
              self.potongan,
              self.gator,
              self.gaber,
              )
              )

#karyawan1 = Gaji("Junaidi Surya", "Ahli Madya", 20, "III", "a")
karyawan2 = Gaji("Andika Surya Saputra", "Ahli Muda", 10, "III", "a")
#karyawan3 = Gaji("Raihan Surya Saputra", "Ahli Utama", 32, "III", "a")
karyawan4 = Gaji("Nabila Nur Assyifa", "Ahli Pertama", 0, "III", "a")
#karyawan1.HitungGaji()
#karyawan1.show_gaji()
print("=" * 40)
karyawan2.HitungGaji()
karyawan2.show_gaji()
print("=" * 40)
#karyawan3.HitungGaji()
#karyawan3.show_gaji()
print("=" * 40)
karyawan4.HitungGaji()
karyawan4.show_gaji()
```

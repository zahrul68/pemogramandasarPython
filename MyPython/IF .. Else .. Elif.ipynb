```python
umur=int(input("Berapa Umur Anda : "))
print("=" * 30)
if umur >= 18:
    print("Kamu Boleh Membuat SIM")
else:
    print("Kamu Belum Boleh Membuat SIM")
```

```python
nama=input("Masukkan Nama          : ")
absen=float(input("Masukkan Jml Pertemuan :"))
tugas=float(input("Masukkan Nilai Tugas   :"))
mid=float(input("Masukkan Nilai Mid     :"))
uas=float(input("Masukkan Ujian Akhir   :"))
print("=" * 40)
nilai=((absen/16)*10) + ((tugas/100)*20) + ((mid/100)*30) + ((uas/100)*40)
if nilai >=80 and nilai <=100 :
    grade="A"
    print("Anda Berhak Mendapatkan Subsidi Quota 1 Mb")
elif nilai >=68 and nilai<=79:
    grade="B"
elif nilai >=56 and nilai <=67:
    grade="C"
elif nilai >=46 and nilai <=55:
    grade="D"
else:
    grade="E"
    print("Anda Harus Banyak Banyak Belajar dan Minum Susu")
print("Nama Mahasiswa  = %s" % nama)
print("Nilai Akhir     = %f" % nilai)
print("Nilai Indeks    = %c" % grade)
```
